###Function for Guassian Fitting

```python
def gaussian_fitting(data):

  # Fit a normal distribution to the data:
  mu, std = norm.fit(data)

  var = np.asarray(data).var()
  mean = np.asarray(data).mean()

  # Plot the histogram.
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.hist(data, bins=25, density=True, alpha=0.6, color='g')
  title = "mean = %.5f,  var = %.5f" % (mean, var)
  plt.title(title)

  # Plot the Gaussian fitting
  x = np.linspace(mu - 5*std, mu + 5*std, 100)
  p = norm.pdf(x, mu, np.sqrt(var))
  plt.subplot(1, 2, 2)
  plt.plot(x, p, 'k', linewidth=2)
  title = "Gaussian Fitting"
  plt.title(title)
  plt.show()
```

#Part B) Translation

```python
df_data_linear = pd.read_csv('linear_translation.csv')
df_data_linear.head()
```

```python
def pdf_and_gaussian(data):

  # Fit a normal distribution to the data:
  mu, std = norm.fit(data)

  var = np.asarray(data).var()
  mean = np.asarray(data).mean()

  # Plot the histogram.
  plt.figure(figsize=(12, 5))
  plt.subplot(1, 3, 1)
  plt.hist(data, bins=25, density=True, alpha=0.6, color='g')
  title = "Histogram : mean = %.5f,  var = %.5f" % (mean, var)
  plt.title(title)
  #Organize Data
  SR_y = pd.Series(data)
  plt.subplot(1,3,2)
  sns.distplot(SR_y, bins=25, color="g")
  title = "PDF of Distribution"
  plt.title(title)
  # Plot the Gaussian fitting
  x = np.linspace(mu - 5*std, mu + 5*std, 100)
  p = norm.pdf(x, mu, np.sqrt(var))
  plt.subplot(1, 3, 3)
  plt.plot(x, p, 'k', linewidth=2)
  title = "Gaussian Fitting"
  plt.title(title)
  plt.show()
```

##2.5 sec

```python
pdf_and_gaussian(df_data_linear["velocity_error_2_5sec"])
```

##5 sec

```python
pdf_and_gaussian(df_data_linear["velocity_error_5sec"])
```

##7.5 sec

```python
pdf_and_gaussian(df_data_linear["velocity_error_7_5sec"])
```
